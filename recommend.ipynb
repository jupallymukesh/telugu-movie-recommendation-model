```python
movie_data=movie_meta_data.merge(ratings,on='movieid')
movie_data.head()
```

```python
item_matrix=movie_data.pivot(index=['user_id'],columns=['movieid'],values='rating').fillna(0)
item_matrix
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
import difflib
```

```python
cf_knn_model=NearestNeighbors(metric='cosine',algorithm='brute',n_neighbors=10,n_jobs=-1)

cf_knn_model.fit(item_matrix)
```

```python
item_matrix = item_matrix.T
```

```python
def get_closest_movie_id(movie_name, movie_meta_data):
    highest_similarity = 0.0
    closest_movie_id = None
    
    for i, title in enumerate(movie_meta_data['Movie']):
        similarity = difflib.SequenceMatcher(None, title.lower(), movie_name.lower()).ratio()
        if similarity > highest_similarity:
            highest_similarity = similarity
            closest_movie_id = movie_meta_data['movieid'][i]
    
    if closest_movie_id is None:
        raise ValueError("No matching or similar movie found")
    
    return closest_movie_id

def movie_recommender_engine(movie_name, matrix, cf_model, n_recs, movie_meta_data):
    # Fit model on matrix
    cf_model.fit(matrix)
    
    # Find closest movie ID
    closest_movie_id = get_closest_movie_id(movie_name, movie_meta_data)

    # Calculate neighbour distances
    distances, indices = cf_model.kneighbors(matrix.iloc[[closest_movie_id]], n_neighbors=n_recs)
    movie_rec_ids = sorted(list(zip(indices.squeeze().tolist(), distances.squeeze().tolist())), key=lambda x: x[1])[::-1]

    # List to store recommendations
    cf_recs = []
    for idx, dist in movie_rec_ids:
        cf_recs.append({'Title': movie_meta_data.loc[idx, 'Movie'], 'Distance': dist})

    # Create DataFrame with recommendations
    df = pd.DataFrame(cf_recs, index=range(1, n_recs + 1))
    
    return df
```

```python
n_recs = 5
movie_recommendations = movie_recommender_engine('don', item_matrix, cf_knn_model, n_recs, movie_meta_data)
movie_recommendations[::-1]
```

```python
#import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
#import numpy as np

# Define a function to calculate mean average precision at k
def mean_average_precision_at_k(actual, predicted, k):
    score = 0.0
    num_hits = 0.0

    for i, p in enumerate(predicted[:k]):
        if p in actual and p not in predicted[:i]:
            num_hits += 1.0
            score += num_hits / (i + 1.0)

    if not actual:
        return 1.0

    return score / min(len(actual), k)

# Example usage for evaluating KNN recommendations
def evaluate_knn(matrix, movie_meta_data, n_recs=5, test_size=0.2):
    # Split the data into training and test sets
    train_indices, test_indices = train_test_split(matrix.index, test_size=test_size, random_state=42)
    train_matrix = matrix.loc[train_indices]
    test_matrix = matrix.loc[test_indices]
    
    # Fit the model on the training set
    cf_knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
    cf_knn_model.fit(train_matrix)
    
    # Generate recommendations for the test set
    map_scores = []
    for test_idx in test_indices:
        distances, indices = cf_knn_model.kneighbors(matrix.loc[[test_idx]], n_neighbors=n_recs + 1)
        recommended_indices = indices.squeeze().tolist()
        
        # Ensure the test instance is not in the recommendations
        if test_idx in recommended_indices:
            recommended_indices.remove(test_idx)
        
        # Truncate to the required number of recommendations
        recommended_indices = recommended_indices[:n_recs]
        
        actual_movies = set(test_indices) - {test_idx}  # Actual movies should exclude the current test movie
        map_score = mean_average_precision_at_k(actual_movies, recommended_indices, n_recs)
        map_scores.append(map_score)
    
    # Calculate mean MAP@k
    mean_map = np.mean(map_scores)
    return mean_map

# Example usage
mean_map_score = evaluate_knn(item_matrix, movie_meta_data, n_recs=5, test_size=0.2)
print(f"Mean Average Precision at 5: {mean_map_score}")
```
